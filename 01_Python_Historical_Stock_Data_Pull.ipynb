# Creates blank dataframe

```python
##Creates test dataframe to index and empty dataframe
df_final = pd.DataFrame(columns=['Open', 'High','Low', "Close", 'Adj Close','Volume', 'SMA_50','Ticker'])
```

# Creates The variables

```python
#Date Logic
now = dt.datetime.now() #Runs until todays date
days = dt.timedelta(1825) #Number of days in the past
start = now - days #Gets The date that is the number of days assigned above
index = 0
```

```python
#Creates Column Names
ma=50 #Moving average amount
smaString="SMA_"+str(ma) #Creates Column Name
alltime=100
year=365 #Creates a variable for a year
highString="Annual_High"#Create Annual High Column
lowString="Annual_Low"#Create Annual High Column
num=0
```

# Pulls data and adds it to df_final

```python
index = 0

for a in list:  
        
    df= pdr.get_data_yahoo(a, start, now) #Create Pandas dateframe with stock info 
        
    
    append = df.index
    number_of_rows = len(append)
    count = str(number_of_rows)

    df[smaString]= df.iloc[:,4].rolling(window=ma).mean() #Creates MA Column
    df=df.iloc[ma:] #Cuts out first 50 columns with no MA
        
    #Assigns itterated value as ticker
    df['Ticker'] = a
    
    #Tracks Index
    index+=1   
        
        
    #Updates final database
    df_final = pd.concat([df_final, df])
```

# Creates the annual fields in a new data frame

```python
#Creates Isolated dateframe
df_year_high = df_final[['Ticker', 'High']]
df_year_low = df_final[['Ticker', 'Low']]

#Creates last 252 items(Trading days a year)
df_year_high = df_year_high.groupby('Ticker').tail(252)
df_year_low = df_year_low.groupby('Ticker').tail(252)

#Creates the max per ticker
df_high = pd.DataFrame(df_year_high.groupby('Ticker').max('High'))
df_low = pd.DataFrame(df_year_low.groupby('Ticker').max('Low'))

#Creates dataframe to isolate high and low for last year.
df_annual = df_high.merge(df_low, how='inner', on='Ticker')

#Renames columns to final column names
df_annual = df_annual.rename(columns={"High": "Annual_High", "Low": "Annual_Low"})
```

# Adds the new fields to the final dataframe.

```python
#Resets Index
df_final = df_final.reset_index()

#Renames columns to final column names
df_final = df_final.rename(columns={"index": "Date"})

#Creates Final dataframe with the annual high and annual low merged in
df_final = df_final.merge(df_annual, how='inner', on='Ticker')

#Reorders columns
df_final = df_final[['Date','Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume','SMA_50','Annual_High','Annual_Low','Ticker']]
```
